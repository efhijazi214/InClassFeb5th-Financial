```python
#Importing the packages we will be using to retrieve stock price data for Amazon and Microsoft
from pandas_datareader import data, wb
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
%matplotlib inline
```

The cells below will retrieve data for Amazon's stock price between Jan 1st and today's date (Feb 15th)

```python
start = datetime.datetime(2022, 1, 1)
end = datetime.datetime(2022, 2, 15)
```

The datareader retrieves Amazon price data from Yahoo Finance.

```python
AMZN = data.DataReader("AMZN", 'yahoo', start, end)
```

```python
AMZN
```

We will use reset index to make data a column and reset our fields and records.

```python
myAMZN = AMZN.reset_index()
```

```python
myAMZN
```

Now we will use plot.line to create a graph of Amazon's year-to-date.

```python
plt.style.use('bmh')
myAMZN.plot.line(x='Date',y='Volume',figsize=(12,3),lw=1)
```

We are plotting a bar chart of the opening for the month of December 2021. First we must define new dates and then plug "Open" into the bar chart function to specify that is the only thing we want. 

```python
start2 = datetime.datetime(2021, 12, 1)
end2 = datetime.datetime(2021, 12, 31)
```

```python
AMZN2 = data.DataReader("AMZN", 'yahoo', start2, end2)
AMZN2["Open"].plot.bar()
```

Now we are going to do different charts with Microsoft price data, first by defining it in the same time period. 

```python
MSFT = data.DataReader("MSFT", 'yahoo', start, end)
```

```python
MSFT
```

Reset Microsoft Index.

```python
myMSFT = MSFT.reset_index()
```

```python
myMSFT
```

Creating the same line chart to study Microsoft's volume year-to-date.

```python
plt.style.use('bmh')
myMSFT.plot.line(x='Date',y='Volume',figsize=(12,3),lw=1)
```

Now we must define a third date and plot a density chart for MSFT's highs and lows  in January 2022.

```python
start3 = datetime.datetime(2022, 1, 1)
end3 = datetime.datetime(2022, 1, 31)
```

```python
MSFT = data.DataReader("MSFT", 'yahoo', start3, end3)
MSFT[['High','Low']].plot.density()
```
